```python
#Dividin x and y / objetive variable
columns = ["TV","Radio", "Newspaper"]
x = df[columns]
x
```

```python
y = df["Sales"]
y
```

```python
#Split
from sklearn.model_selection import train_test_split
#I gonna have four news dataframes
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=42)
#random_state = semilla / seed -> its a way to how the data is gonna be separated
#train 80% / test 20%
```

```python
x_train
```

```python
y_train
```

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
#Review parameter model
regressor.coef_
#B1 B2 B3
#En este caso el radio es el mecanismo que mas ventas me da, ya que su coeficiente es mas grande (suponiendo que estan en las mismas unidades las variables)
```

```python
#Review parameter model
regressor.intercept_
#B0
#El valor de la variable objetivo cuando las otras variables independientes o descritivas son cero
#Este seria el valor de las ventas cuando  el valor del periodico, radio ytv es cero
#Guarda la informacion de las variables que no estamos tomando en cuenta
```

```python
y_pred = regressor.predict(x_test)
```

```python
df = pd.DataFrame({'Real Values' : y_test, 'Predicted Values:' : y_pred})
df
```

```python
#MAE
from sklearn import metrics
print(metrics.mean_absolute_error(y_test, y_pred))
```

```python
#MSE
print(metrics.mean_squared_error(y_test, y_pred))
```

```python
#RMSE
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
#R^2
print(metrics.r2_score(y_test, y_pred))
#Medida de bondad de ajuste del 86 %, este modelo lineal explica el 86$ de la varianza de los datos
```
